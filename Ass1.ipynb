```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

c = np.array([(50, 50), (50, 100), (150, 255), (150, 150)])
t1 = np.linspace(0, c[0, 1], c[0, 0] + 1 - 0).astype('uint8')
print(len(t1))
t2 = np.linspace(c[0, 1] + 1 , c[1, 1], c[1, 0] - c[0, 0]).astype('uint8')
print(len(t2))
t3 = np.linspace(c[1, 1] + 1 , c[2, 1], c[2, 0] - c[1, 0]).astype('uint8')
print(len(t3))
t4 = np.linspace(c[2, 1] + 1 , c[3, 1], c[3, 0] - c[2, 0]).astype('uint8')
print(len(t4))
t5 = np.linspace(c[3, 1] + 1 , 255, 255 - c[3, 0]).astype('uint8')
print(len(t5))

transform = np.concatenate((t1, t2), axis=0).astype('uint8')
transform = np.concatenate((transform, t3), axis=0).astype('uint8')
transform = np.concatenate((transform, t4), axis=0).astype('uint8')
transform = np.concatenate((transform, t5), axis=0).astype('uint8')
print(len(transform))
fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel('Input intensity')
ax.set_ylabel('Output intensity')
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_aspect('equal')
plt.savefig('transform_q1.png')
plt.show()
img_orig = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
cv.namedWindow("Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Image", img_orig)
cv.waitKey(0)
image_transformed = cv.LUT(img_orig, transform)
cv.imshow("Image", image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'highlights_and_shadows.jpg'
image = cv2.imread(image_path)

# Convert the image to the L*a*b* color space
lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Extract the Luminance (L) channel
l_channel = lab_image[:,:,0]

# Gamma correction function
def apply_gamma_correction(image, gamma):
    gamma_corrected = np.power(image / 255.0, gamma) * 255
    return np.uint8(gamma_corrected)

gamma_value = 0.5  # Adjust this value based on your needs
gamma_corrected_l_channel = apply_gamma_correction(l_channel, gamma_value)

# Replace the corrected L channel in the LAB image
lab_image[:,:,0] = gamma_corrected_l_channel

corrected_image = cv2.cvtColor(lab_image, cv2.COLOR_LAB2BGR)
corrected_image1 = cv2.cvtColor(corrected_image, cv2.COLOR_BGR2RGB)

fig,ax=plt.subplots()
# Plot histograms
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(l_channel.ravel(), bins=256, range=(0, 256), color='blue', alpha=0.5, label='Original L channel')
plt.title('Histogram of Original L Channel')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(gamma_corrected_l_channel.ravel(), bins=256, range=(0, 256), color='green', alpha=0.5, label='Gamma Corrected L channel')
plt.title('Histogram of Gamma Corrected L Channel')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
ax.imshow(corrected_image1)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image (replace 'image_path' with the actual path to your image)
image_path = 'F:\Aca sem 5\Image processing\Assignment_1\Image_Processing\spider.png'
image = cv2.imread(image_path)

# Convert the image to the HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Split the HSV image into its planes
hue_plane = hsv_image[:,:,0]
saturation_plane = hsv_image[:,:,1]
value_plane = hsv_image[:,:,2]

# Apply the intensity transformation to the saturation plane
a = 0.7  # Adjust this value for your desired effect
sigma = 70
transformed_saturation_plane = np.minimum(a * saturation_plane + (1 - a) * 128 * np.exp(-(saturation_plane - 128)**2 / (2 * sigma**2)), 255)

# Clip the values of the transformed saturation plane to [0, 255]
transformed_saturation_plane = np.clip(transformed_saturation_plane, 0, 255)

# Convert each plane to np.uint8
hue_plane = hue_plane.astype(np.uint8)
transformed_saturation_plane = transformed_saturation_plane.astype(np.uint8)
value_plane = value_plane.astype(np.uint8)

# Recombine the modified saturation plane with the original hue and value planes
vibrance_enhanced_hsv_image = np.stack([hue_plane, transformed_saturation_plane, value_plane], axis=-1)

# Convert back to BGR color space
vibrance_enhanced_image = cv2.cvtColor(vibrance_enhanced_hsv_image, cv2.COLOR_HSV2BGR)

# Display the images
plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(cv2.cvtColor(vibrance_enhanced_image, cv2.COLOR_BGR2RGB))
plt.title('Vibrance-Enhanced Image')
plt.axis('off')

plt.subplot(133)
x = np.linspace(0, 255, 256)
y = np.minimum(a * x + (1 - a) * 128 * np.exp(-(x - 128)**2 / (2 * sigma**2)), 255)
plt.plot(x, y)
plt.title('Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Transformed Intensity')

plt.tight_layout()
plt.show()

```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
img = cv.imread('shells.jpg', cv.IMREAD_GRAYSCALE)

# Compute and plot the cumulative distribution function (CDF) of the original image
hist, bins = np.histogram(img.ravel(), 256, [0, 256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()
plt.plot(cdf_normalized, color='b')
plt.hist(img.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.legend(('cdf', 'histogram'), loc='upper left')
plt.title('Histogram of the Original Image')
plt.show()

# Apply histogram equalization to the image
equ = cv.equalizeHist(img)

# Compute and plot the CDF of the equalized image
hist, bins = np.histogram(equ.ravel(), 256, [0, 256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()
plt.plot(cdf_normalized, color='b')
plt.hist(equ.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.legend(('cdf', 'histogram'), loc='upper left')
plt.title('Histogram of the Equalized Image')
plt.show()

# Display the original and equalized images side by side
res = np.hstack((img, equ))
plt.axis('off')
plt.imshow(res, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image (replace 'image_path' with the actual path to your image)
image_path = 'jeniffer.jpg'
image = cv2.imread(image_path)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Split the HSV image into its planes
saturation_plane = hsv_image[:,:,1]

# Display the saturation plane
plt.figure(figsize=(6, 6))
plt.imshow(saturation_plane, cmap='gray')
plt.title('Saturation Plane')
plt.axis('off')
plt.show()

# Calculate and plot the histogram of the saturation plane
hist, bins = np.histogram(saturation_plane.ravel(), 256, [0, 256])
plt.plot(hist, color='b')
plt.xlim([0, 256])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Histogram of Saturation Plane')
plt.show()

# Based on visual inspection and histogram analysis, choose a threshold value
threshold_value = 120  # You should adjust this value based on your observations

# Threshold the saturation plane to create a binary mask
_, thresholded_mask = cv2.threshold(saturation_plane, threshold_value, 255, cv2.THRESH_BINARY)

# Convert the thresholded mask to a binary mask (0 and 255)
binary_mask = np.where(thresholded_mask > 0, 255, 0).astype(np.uint8)

# Extract the foreground using cv.bitwise_and
foreground = cv2.bitwise_and(image, image, mask=binary_mask)

# Compute histogram of the foreground
foreground_histogram = cv2.calcHist([foreground], [0], None, [256], [0, 256])

# Compute cumulative sum of the histogram
cumulative_histogram = np.cumsum(foreground_histogram)

# Normalize the cumulative histogram
cumulative_histogram_normalized = cumulative_histogram / cumulative_histogram[-1]

# Apply histogram equalization to the foreground
equalized_foreground = np.interp(foreground[:,:,0], np.arange(256), cumulative_histogram_normalized).astype(np.uint8)

# Extract background using the inverse of the binary mask
background_mask = cv2.bitwise_not(binary_mask)
background = cv2.bitwise_and(image, image, mask=background_mask)

# Combine the equalized foreground and background
result_image = cv2.add(background, cv2.cvtColor(equalized_foreground, cv2.COLOR_GRAY2BGR))

# Display the original image, equalized foreground, and result
plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(cv2.cvtColor(equalized_foreground, cv2.COLOR_BGR2RGB))
plt.title('Equalized Foreground')
plt.axis('off')

plt.subplot(133)
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.title('Result with Equalized Foreground')
plt.axis('off')

plt.tight_layout()
plt.show()

```

```python
def sobel_filter(image):
    kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
    
    sobel_x = cv2.filter2D(image, cv2.CV_64F, kernel_x)
    sobel_y = cv2.filter2D(image, cv2.CV_64F, kernel_y)
    
    gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    
    return gradient_magnitude

# Load the image and apply Sobel filtering using your function
image = cv2.imread('F:\Aca sem 5\Image processing\Assignment_1\Image_Processing\einstein.png', cv2.IMREAD_GRAYSCALE)
gradient_magnitude_custom = sobel_filter(image)

# Display the original image and the gradient magnitude
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(gradient_magnitude_custom, cmap='gray')
plt.title('Gradient Magnitude (Custom Sobel)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

# Define the zoom factor and scale-up factor for testing
zoom_factor = 4
scale_factor = 1 / zoom_factor

# Load the original and zoomed-out images
original_image = cv2.imread('F:\Aca sem 5\Image processing\Assignment_1\Image_Processing\zooming\im01small.png', cv2.IMREAD_GRAYSCALE)
zoomed_out_image = cv2.imread('F:\Aca sem 5\Image processing\Assignment_1\Image_Processing\zooming\im01.png', cv2.IMREAD_GRAYSCALE)

# Define a function for nearest-neighbor zooming
def nearest_neighbor_zoom(image, factor):
    new_height = int(image.shape[0] * factor)
    new_width = int(image.shape[1] * factor)
    zoomed_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_NEAREST)
    return zoomed_image

# Define a function for bilinear interpolation zooming
def bilinear_interpolation_zoom(image, factor):
    new_height = int(image.shape[0] * factor)
    new_width = int(image.shape[1] * factor)
    zoomed_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
    return zoomed_image

# Zoom the original images using both methods
zoomed_nn = nearest_neighbor_zoom(original_image, zoom_factor)
zoomed_bilinear = bilinear_interpolation_zoom(original_image, zoom_factor)

# Calculate the normalized sum of squared difference (SSD)
ssd_nn = np.sum((zoomed_nn - zoomed_out_image) ** 2) / np.prod(zoomed_out_image.shape)
ssd_bilinear = np.sum((zoomed_bilinear - zoomed_out_image) ** 2) / np.prod(zoomed_out_image.shape)

# Print the SSD values
print(f"Nearest-Neighbor SSD: {ssd_nn:.2f}")
print(f"Bilinear Interpolation SSD: {ssd_bilinear:.2f}")

# Display the zoomed images
cv2.imshow('Original', original_image)
cv2.imshow('Zoomed Nearest-Neighbor', zoomed_nn)
cv2.imshow('Zoomed Bilinear Interpolation', zoomed_bilinear)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
